### Leitura do dataset

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('AH_Provisional_Diabetes_Death_Counts_for_2020.csv')
dataset
```

### Ajuste dos tipos das variáveis

```python
dataset.dtypes
```

```python
dataset['Data as of'] = dataset['Data as of'].astype('category')
dataset['Date_Of_Death_Year'] = dataset['Date_Of_Death_Year'].astype('category')
dataset['Date_Of_Death_Month'] = dataset['Date_Of_Death_Month'].astype('category')
dataset['AgeGroup'] = dataset['AgeGroup'].astype('category')
dataset['Sex'] = dataset['Sex'].astype('category')
dataset.dtypes
```

```python
print(dataset['Data as of'].cat.categories)
print(dataset['Date_Of_Death_Year'].cat.categories)
print(dataset['Date_Of_Death_Month'].cat.categories)
print(dataset['AgeGroup'].cat.categories)
print(dataset['Sex'].cat.categories)
```

**Remoção das colunas Data as of e Date_Of_Death_Year, pois estas não variam ao longo do dataset**

```python
dataset = dataset.iloc[:,2:] 
dataset.head()
```

### Verificação de instâncias vazias

```python
dataset.isnull().sum()
```

**Localização da instância nula**

```python
dataset[dataset['AgeGroup'].isnull()]
```

```python
dataset.loc[125:129]
```

**Essa variável vazia pode ser interpretada como similar á cateogria '*Unknown age*', portanto, o valores dessas duas linhas serão somados e a linha contendo o NaN será removida**

```python
dataset.iloc[126,3:-1] = dataset.iloc[126,3:-1] + dataset.iloc[127,3:-1]
dataset = dataset.drop(127)
dataset = dataset.reset_index(drop = True)
dataset.loc[125:129]
```

### Verificação de duplicatas

```python
dataset[dataset.duplicated()]
```

**Não existem linhas duplicadas**

```python
dataset.T.duplicated()
```

**Uma coluna foi identificada como duplicada. Esta será removida.**

```python
dataset = dataset.T.drop_duplicates().T
dataset.head()
```

```python
dataset.dtypes
```

```python
for atributo in dataset.columns:
    if atributo in ['Date_Of_Death_Month','AgeGroup','Sex']:
        dataset[atributo] = dataset[atributo].astype('category')
    else:
        dataset[atributo] = dataset[atributo].astype('int64')
dataset.dtypes
```

### Detecção de Outliers

#### **Método do Z-score robusto**

**Verificação de outliers na variável de mortes por COVID-19 ('COVID19')**

Distribuição da variável de mortes por COVID-19 ('COVID19')

```python
dataset['COVID19'].plot.box()
```

```python
dataset['COVID19'].plot.hist()
```

Aproximação da curva normal pelo log10

```python
dataset['COVID19_log'] = dataset['COVID19'].apply(lambda x:np.log10(x+1))
dataset['COVID19_log'].plot.box()
```

```python
dataset['COVID19_log'].plot.hist()
```

Busca pelos outliers com Mi > 3.5

```python
mad = np.abs(dataset['COVID19_log'] - dataset['COVID19_log'].median()).median()*(1/0.6745)
(np.abs(dataset['COVID19_log']-dataset['COVID19_log'].median())/mad).hist()
dataset[np.abs(dataset['COVID19_log']-dataset['COVID19_log'].median())/mad > 3.5]
```

Não foram detectados outliers na variável de mortes por Covid-19

**Verificação de outliers na variável de mortes causadas e associadas à Diabetes ('Diabetes.mc')**

Distribuição da variável de mortes mortes causadas e associadas à Diabetes ('Diabetes.mc')

```python
dataset['Diabetes.mc'].plot.box()
```

```python
dataset['Diabetes.mc'].plot.hist()
```

A distribução não apresenta outliers no boxplot, além de ser aproximadamente normal, sem a necessidade de ajuste por log

Busca pelos outliers com Mi > 3.5

```python
mad = np.abs(dataset['Diabetes.mc'] - dataset['Diabetes.mc'].median()).median()*(1/0.6745)
(np.abs(dataset['Diabetes.mc']-dataset['Diabetes.mc'].median())/mad).hist()
dataset[np.abs(dataset['Diabetes.mc']-dataset['Diabetes.mc'].median())/mad > 3.5]
```

Não foram detectados outliers na variável de mortes causadas e associadas à Diabetes ('Diabetes.mc')

**Verificação de outliers na variável da soma de mortes causadas e associadas à Diabetes ('Diabetes.mc') e à Covid-19 ('COVID19')**

Distribuição da variável de mortes por COVID-19 mais mortes causadas ou associadas à diabetes (COVID19_Diabetes_deaths)

```python
dataset['COVID19_Diabetes_deaths'] = dataset['COVID19'] + dataset['Diabetes.mc']
dataset['COVID19_Diabetes_deaths'].plot.box()
```

```python
dataset['COVID19_Diabetes_deaths'].plot.hist()
```

Aproximação da curva normal pelo log10

```python
dataset['COVID19_Diabetes_deaths_log'] = dataset['COVID19_Diabetes_deaths'].apply(lambda x:np.log10(x))
dataset['COVID19_Diabetes_deaths_log'].plot.box()
```

```python
dataset['COVID19_Diabetes_deaths_log'].plot.hist()
```

Busca pelos outliers com Mi > 3.5

```python
mad = np.abs(dataset['COVID19_Diabetes_deaths_log'] - dataset['COVID19_Diabetes_deaths_log'].median()).median()*(1/0.6745)
(np.abs(dataset['COVID19_Diabetes_deaths_log']-dataset['COVID19_Diabetes_deaths_log'].median())/mad).hist()
dataset[np.abs(dataset['COVID19_Diabetes_deaths_log']-dataset['COVID19_Diabetes_deaths_log'].median())/mad > 3.5]
```

Os outliers encontrados nas mortes por COVID19 são associados à baixa mortalidade jovens menores de 18 anos e à catgoria residual de idade desconhecida, a qual reporta poucos casos

#### **Método de Tukey**

**Verificação de outliers na variável de mortes por COVID-19 ('COVID19')**

```python
q1 = dataset['COVID19_log'].quantile(q=0.25)
q3 = dataset['COVID19_log'].quantile(q=0.75)
iqr = q3 - q1
print(iqr)
```

Outliers inferiores

```python
dataset[dataset['COVID19_log'] < (q1-1.5*iqr)]
```

Outliers superiores

```python
dataset[dataset['COVID19_log'] > (q3+1.5*iqr)]
```

**Verificação de outliers na variável de mortes causadas e associadas à Diabetes ('Diabetes.mc')**

```python
q1 = dataset['Diabetes.mc'].quantile(q=0.25)
q3 = dataset['Diabetes.mc'].quantile(q=0.75)
iqr = q3 - q1
print(iqr)
```

Outliers inferiores

```python
dataset[dataset['Diabetes.mc'] < (q1-1.5*iqr)]
```

Outliers superiores

```python
dataset[dataset['Diabetes.mc'] > (q3+1.5*iqr)]
```

**Verificação de outliers na variável de mortes causadas e associadas à Diabetes ('Diabetes.mc') e à Covid-19 ('COVID19')**

```python
q1 = dataset['COVID19_Diabetes_deaths_log'].quantile(q=0.25)
q3 = dataset['COVID19_Diabetes_deaths_log'].quantile(q=0.75)
iqr = q3 - q1
print(iqr)
```

Outliers inferiores

```python
dataset[dataset['COVID19_Diabetes_deaths_log'] < (q1-1.5*iqr)]
```

Outliers superiores

```python
dataset[dataset['COVID19_Diabetes_deaths_log'] > (q3+1.5*iqr)]
```

Os outliers estão associados à baixa mortalidade de jovens menores de 18 anos nos primeiros meses de 2020. A outra categoria associada é a categoria residual de idade desconhecida.

### Análise de mortes por mês no total

**Verificação de sobreposição de faixa etária.**

```python
dataset['AgeGroup'].value_counts().sort_index()
```

**Remoção das faixas de idade '*50-64 years*, *65-74 years*, *75-84 years*, *85+ years*' do sub-dataset de mortes por ano para evitar a sobrepoisção com outras duas categorias, o que resultaria na cotabilização de mortes duplicadas.**

```python
death_by_month = dataset.drop(dataset[(dataset['AgeGroup'] == '50-64 years')|
       (dataset['AgeGroup'] == '65-74 years')|
       (dataset['AgeGroup'] == '75-84 years')|
       (dataset['AgeGroup'] == '85+ years')].index.tolist())
death_by_month = death_by_month.reset_index(drop = True)
death_by_month
```

```python
death_by_month = death_by_month.groupby(['Date_Of_Death_Month']).sum()
death_by_month
```

```python
death_by_month.plot.bar(y = ['COVID19','Diabetes.uc'])
```

```python
death_by_month.plot.bar(y = ['COVID19','Diabetes.mc'])
```

```python
death_by_month.plot.scatter(x ='COVID19',y = 'Diabetes.uc')
```

```python
death_by_month[['COVID19','Diabetes.uc','Diabetes.mc']].corr().style.background_gradient(cmap='coolwarm', axis=None)
```

```python
death_by_month.plot.scatter(x ='COVID19',y = 'Diabetes.mc')
```

**Mortes por mês na escala do logarítimo de 10**

```python
death_by_month_log = death_by_month.apply(lambda x:np.log10(x+1))
death_by_month_log.plot.bar(y = ['COVID19','Diabetes.mc'])
```

**Quais faotres de morte associados ao COVID19 possuem maior propoção em relação ao total de mortes?**

```python
death_by_month[['C19PlusDiabetes','C19PlusHypertensiveDiseases',
               'C19PlusMajorCardiovascularDiseases',
               'C19PlusChronicLowerRespiratoryDisease',
               'C19PlusKidneyDisease',
               'C19PlusChronicLiverDiseaseAndCirrhosis',
               'C19PlusObesity']].sum().apply(lambda x: x/death_by_month['COVID19'].sum()).sort_values(ascending=False)
```
